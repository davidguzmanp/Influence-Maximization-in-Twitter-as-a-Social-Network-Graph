```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import figure
```

```python
g = nx.read_gml('PB2020.gml')
g = g.reverse()
```

```python
#print number of nodes 
print("Number of nodes: ", g.number_of_nodes())
#print number of edges
print("Number of edges: ", g.number_of_edges())
```

```python
deg = [g.degree(n) for n in g.nodes()]
avg_deg = np.mean(deg)
print("Average degree: ", avg_deg)
```

```python
#increase plot size
plt.figure(figsize=(10,6), dpi=110)
my_bins = np.logspace(0,np.log10(max(deg) +1), 30) 
plt.hist(deg, bins=my_bins, density = True, color = "lightblue", ec="black")
plt.title("Degree distribution Twitter")
plt.ylabel("Probability")
plt.xlabel("Degree") 
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# compute the degree clustering coefficient
clustering = nx.clustering(g)
print("Average clustering coefficient: ", np.mean(list(clustering.values())))
```
