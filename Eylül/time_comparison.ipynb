```python
#read all JSON files and save them in a list
def read_json_files():

    import json
    import os
    import glob

    path = r'C:\Users\eylul\Desktop\2022-Fall\Network Science\Final\Network-Science-final-Maximization\Eyl√ºl' # use your path
    all_files = glob.glob(os.path.join(path, "*.json"))     # advisable to use os.path.join as this makes concatenation OS independent

    json_list = []
    for file_ in all_files:
        with open(file_, "r", encoding="utf8") as f:
            json_list.append(json.load(f))

    return json_list
```

```python
a = read_json_files()
```

```python
# save index 2 of all items in the list
def save_index_2():
    list_ = []
    for i in range(len(a)):
        list_.append(a[i][2])
    return list_

def save_index_1():
    list_ = []
    for i in range(len(a)):
        list_.append(a[i][1])
    return list_
```

```python
times = save_index_2()
spreads = save_index_1()
```
